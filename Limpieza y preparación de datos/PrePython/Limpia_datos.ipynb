```python
import os
import pandas as pd
import re

# Ruta de la carpeta que quieres leer
ruta_carpeta = "C:/Users/victo/OneDrive/Escritorio/PROY II/CCAA_3"

# Obtener una lista de los archivos en la carpeta
archivos = os.listdir(ruta_carpeta)

# Crear un DataFrame vacío para almacenar los datos
dataframe = pd.DataFrame()
tipo = ["_Generación", "_Demanda"]
# Iterar sobre la lista de archivos
for archivo in archivos:
    ruta_archivo = os.path.join(ruta_carpeta, archivo)
    # Leer el archivo CSV y agregar una columna nueva con el nombre del archivo
    df = pd.read_csv(ruta_archivo)
    archivo = re.findall(r'([A-Z]\w+[a-z].+)_genvsdem', archivo)[0]
    df['Comunidad Autonoma'] = archivo
    
    # Concatenar el DataFrame actual al DataFrame principal
    dataframe = pd.concat([dataframe, df], ignore_index=True)

#Concatenamos la columna de Comunidad Autónoma (donde pone el nombre de cada Comunidad Autónoma) con el tipo de dato "Unnamed: 0" (Generación/Demanda)
dataframe['Comunidad Autonoma'] = dataframe['Comunidad Autonoma'].str.cat(dataframe['Unnamed: 0'], sep='_')

# Elminamos la primera columna para que nos sea más sencillo al análisis
dataframe = dataframe.drop(columns=['Unnamed: 0'])
```

```python
dataframe.head()
```

```python
# Trasponemos el dataframe para que tenga el aspecto de los que tratamos en R
genvsdem = dataframe.set_index('Comunidad Autonoma').T

# Reestablecemos el índice para tener los meses en una columna (que se llamará "X")
genvsdem.reset_index(inplace=True)
genvsdem.rename(columns={'index': 'X'}, inplace=True)
#Ponemos a nula o None la columna con los índices, porque al transponerlo se nos quedaba con el nombre de "Comunidad Autonoma", para evitar mal
#interpretados
genvsdem.columns.name = None
```
